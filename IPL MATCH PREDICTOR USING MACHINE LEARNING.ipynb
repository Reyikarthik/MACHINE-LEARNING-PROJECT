```python
teams=['Sunrisers Hyderabad', 'Mumbai Indians',
       'Royal Challengers Bangalore',
       'Kolkata Knight Riders', 'Kings XI Punjab',
       'Chennai Super Kings', 'Rajasthan Royals',
       'Delhi Capitals']
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df=match_df[['city','winner','total_runs','match_id']]
```

```python
match_df
```

```python
delivery_df=match_df.merge(delivery,on='match_id')
```

```python
delivery_df
```

```python
delivery_df=delivery_df[delivery_df['inning']==2]
```

```python
delivery_df
```

```python
delivery_df['current_score']=delivery_df.groupby('match_id')['total_runs_y'].cumsum()
```

```python
delivery_df
```

```python
delivery_df['runs_left']=delivery_df['total_runs_x']-delivery_df['current_score']
```

```python
delivery_df
```

```python
delivery_df['balls_left']=126-delivery_df['over']*6-delivery_df['ball']
```

```python
delivery_df
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
# Group by match_id (and optionally innings, if available) and compute cumulative sum
delivery_df['cumulative_wickets'] = delivery_df.groupby(['match_id'])['player_dismissed'].cumsum()

# Ensure cumulative dismissals do not exceed 10
delivery_df['cumulative_wickets'] = delivery_df['cumulative_wickets'].clip(upper=10)

# Calculate remaining wickets
delivery_df['wickets'] = 10 - delivery_df['cumulative_wickets']
```

```python
delivery_df.tail(10)
```

```python
delivery_df=delivery_df.rename(columns={'wickets':'wickets_left'})
```

```python
delivery_df.head(30)
```

```python
# CALCULATING THE CURRENT RUNRATE
#CRR= runs/overs
```

```python
delivery_df['CRR']=(delivery_df['current_score']*6/(120-delivery_df['balls_left']))
```

```python
delivery_df
```

```python
# CALCULATING THE REQUIRED RUN RATE
```

```python
delivery_df['RRR']=(delivery_df['runs_left']*6/(delivery_df['balls_left']))
```

```python
delivery_df
```

```python
def result(row):
    if row['batting_team']==row['winner'] :
        return 1
    else:
        return 0
```

```python
delivery_df['result'] = delivery_df.apply(result,axis=1)
```

```python
final_df=delivery_df[['bowling_team','batting_team','city','runs_left','balls_left','wickets_left','CRR',
       'RRR', 'result']]
```

```python
final_df.sample()
```

```python
final_df.dropna(inplace=True) 
```

```python
# MODEL TRAING
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=final_df.iloc[:,0:8]
y=final_df.iloc[:,-1]
```

```python
x.head()
```

```python
y.sample(5)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
transformer=ColumnTransformer(
    transformers=[
        ('trf1',OneHotEncoder(sparse_output=False,drop='first'),['bowling_team','batting_team','city'])],
    remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe=Pipeline(steps=[
    ('step1',transformer),
    ('step2',LogisticRegression(solver='liblinear'))])
```

```python
final_df.isnull().sum()
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)*100
```

```python
# FINDING THE WINNING AND LOSSING PERCENTAGE OF THE CHASING TEAM AFTER EACH BALL IN THE MATCH
```

```python
def match_progression(x_df,match_id,pipe):
    match = x_df[x_df['match_id'] == match_id]
    match = match[(match['ball'] == 6)]
    temp_df = match[['batting_team','bowling_team','city','runs_left','balls_left','wickets_left','total_runs_x','CRR','RRR']].dropna()
    temp_df = temp_df[temp_df['balls_left'] != 0]
    result = pipe.predict_proba(temp_df)
    temp_df['lose'] = np.round(result.T[0]*100,1)
    temp_df['win'] = np.round(result.T[1]*100,1)
    temp_df['end_of_over'] = range(1,temp_df.shape[0]+1)
    
    target = temp_df['total_runs_x'].values[0]
    runs = list(temp_df['runs_left'].values)
    new_runs = runs[:]
    runs.insert(0,target)
    temp_df['runs_after_over'] = np.array(runs)[:-1] - np.array(new_runs)
    wickets = list(temp_df['wickets_left'].values)
    new_wickets = wickets[:]
    new_wickets.insert(0,10)
    wickets.append(0)
    w = np.array(wickets)
    nw = np.array(new_wickets)
    temp_df['wickets_in_over'] = (nw - w)[0:temp_df.shape[0]]
    
    print("Target-",target)
    temp_df = temp_df[['end_of_over','runs_after_over','wickets_in_over','lose','win']]
    return temp_df,target
    
```

```python
temp_df,target = match_progression(delivery_df,12,pipe)
temp_df
```

```python
# CREATING THE PLOT WHICH SHOWS THE MATCH PROGRESSION 
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(18,8))
plt.plot(temp_df['end_of_over'],temp_df['wickets_in_over'],color='yellow',linewidth=3)
plt.plot(temp_df['end_of_over'],temp_df['win'],color='#00a65a',linewidth=4)
plt.plot(temp_df['end_of_over'],temp_df['lose'],color='red',linewidth=4)
plt.bar(temp_df['end_of_over'],temp_df['runs_after_over'])
plt.title('Target-' + str(target))
```
